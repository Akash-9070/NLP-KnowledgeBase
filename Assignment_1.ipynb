```python
import requests
import time

# Base URL of the API endpoint
base_url = "https://api.themoviedb.org/3/movie/top_rated"
api_key = "8265bd1679663a7ea12ac168da84d2e8"

# List to store the fetched data from all pages
all_movies = []

# Loop through pages 1 to 471
for page in range(1, 51):
    # Construct the URL for the current page
    url = f"{base_url}?api_key={api_key}&language=en-US&page={page}"

    # Make a GET request to the API
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON data
        data = response.json()

        # Append the results to the all_movies list
        all_movies.extend(data['results'])

        # Optional: Print progress
        print(f"Fetched data for page {page}")

        # To prevent overwhelming the server, add a short delay
        time.sleep(0.5)
    else:
        print(f"Failed to fetch data for page {page}. HTTP Status code: {response.status_code}")
        break

# Check the number of movies fetched
print(f"Total movies fetched: {len(all_movies)}")

# (Optional) Save the data to a file or process it further

# -------------------------------------------------------------------------------------


# Assuming `all_movies` is the list of dictionaries with movie data

# Initialize empty lists for titles, overviews, and genre_ids
titles = []
overviews = []
genre = []

# Loop through each movie in the list
for movie in all_movies:
    titles.append(movie['original_title'])
    overviews.append(movie['overview'])
    genre.append(movie['genre_ids'])

# # Print the results (optional)
# print("Titles:", titles)
# print("Overviews:", overviews)
# print("Genre IDs:", genre)

# -------------------------------------------------------------------------------------
# The mapping of genre IDs to genre names
genre_mapping = {
    28: "Action",
    12: "Adventure",
    16: "Animation",
    35: "Comedy",
    80: "Crime",
    99: "Documentary",
    18: "Drama",
    10751: "Family",
    14: "Fantasy",
    36: "History",
    27: "Horror",
    10402: "Music",
    9648: "Mystery",
    10749: "Romance",
    878: "Science Fiction",
    10770: "TV Movie",
    53: "Thriller",
    10752: "War",
    37: "Western"
}


# Replace genre IDs with names
genre_names_list = [[genre_mapping[genre_id] for genre_id in genre_ids] for genre_ids in genre]

# Print the updated list
# for genre_names in genre_names_list:
    # print(genre_names)
genre = genre_names_list

# ---------------------------------------------------------------------------------
import csv
import pandas as pd

# Name of the CSV file
csv_file = "movies.csv"

# Open the file in write mode
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    # Write the header
    writer.writerow(["Title", "Overview", "Genres"])

    # Write the rows
    for title, overview, genre_list in zip(titles, overviews, genre):
        # Convert the list of genres into a comma-separated string
        genres_str = ", ".join(genre_list)
        # Write the data as a row
        writer.writerow([title, overview, genres_str])

print(f"CSV file '{csv_file}' created successfully.")

final_data = pd.read_csv("movies.csv")

```

```python
data = final_data


# TEXT PREPROCESSING

# 1) Lowercasing..
data["Overview"] = data["Overview"].str.lower()

# # 2) Remove html tags..
import re
def remove_html_tags(text):
  pattern  = re.compile('<.*?>')
  return pattern.sub(r'',text)

data["Overview"] = data["Overview"].apply(remove_html_tags)

# 3) Remove URLs..

def remove_url(text):
  pattern = re.compile(r'http[s]?://\S+')
  return pattern.sub(r'',text)

data["Overview"] = data["Overview"].apply(remove_url)

# # 4) Remove Punctuations

import string ,time

exclude = string.punctuation

def remove_punc(text):
  return text.translate(str.maketrans('','',exclude))

data["Overview"] = data["Overview"].apply(remove_punc)

```
